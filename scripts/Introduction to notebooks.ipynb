# Read from Unity Catalog

### Databricks allows us to write our queries directly in SQL

```python
%sql
SELECT * FROM tesco_bank_training.user_david_herbert.uk_financial_dataset
```

### We can also write our queries directly in `python`

```python
display("tesco_bank_training.user_david_herbert.uk_financial_dataset")
```

### Creating a dataframe in Pyspark

```python
fin_dataset = spark.table("tesco_bank_training.user_david_herbert.uk_financial_dataset")
```

### This can also be done in spark sql

```python
fin_dataset_sparksql = spark.sql("select * from tesco_bank_training.user_david_herbert.uk_financial_dataset")
```

### How do we write a unity catalog table in pyspark

```python
fin_dataset_sparksql.saveTableAs("tesco_bank_training.user_david_herbert.uk_financial_dataset")
```

### This can also be done using spark sql

```python
%sql
CREATE TABLE tesco_bank_training.user_david_herbert.uk_financial_dataset_updated
AS SELECT * FROM tesco_bank_training.user_david_herbert.uk_financial_dataset
```

### Testing 

```python
from pyspark.testing import assertDataFrameEqual
assertDataFrameEqual(fin_dataset, fin_dataset_sparksql)
```
